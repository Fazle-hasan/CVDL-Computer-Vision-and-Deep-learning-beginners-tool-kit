```python
#Importing the open cv
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#To read image cv2.imread function
image = cv2.imread("Lenna_(test_image).png", cv2.IMREAD_COLOR)
#Creating GUI window to display an image on screen
cv2_imshow(image)
```

Finding Image Data Type

```python
print("The data type of the image is, ", image.dtype)
```

```python
#creating GUI window to display the image
cv2_imshow(image)
#for removing/deleting created GUI window from screen and memory
cv2.destroyAllWindows()
```

```python
rows,cols = image.shape[:2]
#(col/2,rows/2) is the center of rotation for the image
# M is the cordinates of the center
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(image,M,(cols,rows))
plt.imshow(dst)
```

```python
#converting image to Gray scale
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print(gray_image)
#plotting the grayscale image
plt.imshow(gray_image)
```

```python
#converting image to HSV format
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
#print(hsv_image)
#plotting the HSV image
plt.imshow(hsv_image)
```

```python
# Canny Edge Detection
edges = cv2.Canny(image=image, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
cv2_imshow(edges)
```
